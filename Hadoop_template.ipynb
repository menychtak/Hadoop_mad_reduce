mapper.py

*   The mapper reads data from STDIN, processes it by:

Converting all words to lowercase to ensure case insensitivity.
Removing punctuation and stopwords to focus only on meaningful words.
Sorting the letters of each word to create a unique key that groups anagrams together.
Outputting a list of lines mapping sorted anagram keys to the count 1 to STDOUT.
Each word is transformed into a standardized form, allowing anagrams to be grouped under the same key.

reducer.py


*   The reducer reads the sorted anagram keys and their occurrences from STDIN, then:

Aggregates the counts of each anagram key to determine how many words share the same set of letters.
Outputs the final count of each unique anagram group to STDOUT.
This results in a summary of how many words in the dataset belong to each anagram group.

```
#CONTENTS OF mapper.py

import sys
import io
import re
import nltk

nltk.download('stopwords', quiet=True)
from nltk.corpus import stopwords

# Define stop words and punctuations
stop_words = set(stopwords.words('english'))
punctuations = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''

# Read input from STDIN
input_stream = io.TextIOWrapper(sys.stdin.buffer, encoding='latin1')

for line in input_stream:
    line = line.strip().lower()  # Convert to lowercase
    line = re.sub(r'[^\w\s]', '', line)  # Remove punctuation using regex

    # Remove any remaining punctuations manually
    for x in punctuations:
        line = line.replace(x, " ")

    words = line.split()  # Split into words

    for word in words:
        if word not in stop_words:  # Ignore stop words
            sorted_word = "".join(sorted(word))  # Sort letters to create anagram key
            print(f"{sorted_word}\t1")  # Emit (sorted_word, 1)


# This is formatted as code
```




```

#CONTENTS OF reducer.py

import sys

current_key = None
current_count = 0

# Read input from STDIN
for line in sys.stdin:
    line = line.strip()  # Remove leading/trailing spaces

    # Parse input (sorted_word, count)
    try:
        key, value = line.split("\t")
        value = int(value)  # Convert count to integer
    except ValueError:
        continue  # Ignore lines with incorrect format

    # If the same key (anagram group), accumulate count
    if current_key == key:
        current_count += value
    else:
        if current_key:
            print(f"{current_key}\t{current_count}")  # Print previous key's count
        current_key = key
        current_count = value  # Reset count for new key

# Output the last key if needed
if current_key:
    print(f"{current_key}\t{current_count}")


```
